```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
### RUN FOR BOTH MMSE AND BADL USING BEST MODELS' MSE ###

mse_model =   
mse_baseline =   
sample_sizes = np.arange(20, 100, 10)  
n_simulations = 1000  
alpha = 0.05  # Significance level

powers = []
ci_lower = []
ci_upper = []

for n in sample_sizes:
    successful_tests = []
    
    for _ in range(n_simulations):
        model_errors = np.random.normal(loc=0, scale=np.sqrt(mse_model), size=n)
        baseline_errors = np.random.normal(loc=0, scale=np.sqrt(mse_baseline), size=n)
        
        mse_model_sim = np.mean(model_errors**2)
        mse_baseline_sim = np.mean(baseline_errors**2)
        
        t_stat, p_value = stats.ttest_rel(model_errors**2, baseline_errors**2)
        
        success = int((p_value < alpha) and (mse_model_sim < mse_baseline_sim))
        successful_tests.append(success)
    
    power_estimate = np.mean(successful_tests)
    powers.append(power_estimate)

    ci = stats.binom.interval(0.95, n_simulations, power_estimate, loc=0)
    lower = ci[0] / n_simulations
    upper = ci[1] / n_simulations
    ci_lower.append(lower)
    ci_upper.append(upper)

plt.figure(figsize=(10,6))
plt.plot(sample_sizes, powers, marker='o', color='#803408', label='Estimated power')
plt.fill_between(sample_sizes, ci_lower, ci_upper, color='#803408', alpha=0.2, label='95% CI')
plt.axhline(0.8, color='red', linestyle='--', label='80% power')
plt.xlabel('Sample size')
plt.ylabel('Power')
plt.title('Power curve: MMSE Model')
plt.legend()

plt.grid(color='lightgrey', linestyle='--', linewidth=1)
plt.show()
```
